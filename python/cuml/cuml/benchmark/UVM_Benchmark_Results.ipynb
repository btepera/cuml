```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
uvm_df = pd.read_json(path_or_buf="kmeans-benchmark-results-uvm.jsonl", lines=True)
no_uvm_df = pd.read_json(path_or_buf="kmeans-benchmark-results-no-uvm.jsonl", lines=True)
no_uvm_df["managed_memory"]=0
df = pd.concat([uvm_df, no_uvm_df])
df["Data Size (GB)"] = df["nrows"]*df["ncols"]*4/1.024E9
pivot = df.pivot(index='Data Size (GB)', columns='managed_memory', values='fit_time')
plot = pivot.plot(marker='o')
plot.set_ylabel('Time (s)')
plot.axvline(x = 40, color = 'k')
plt.legend( loc = 'lower right', title="UVM")
plt.title("KMeans")
plt.show()
```
